# Lectura de datos y creación de los DataFrames

```python
ruta_tienda1='/content/drive/MyDrive/Udemy- Python para data science/dia 7/Datos_Ventas_Tienda.csv'
```

```python
ruta_tienda2='/content/drive/MyDrive/Udemy- Python para data science/dia 7/Datos_Ventas_Tienda2.csv'
```

```python
df=pd.read_csv(ruta_tienda1)
df
```

```python
df2=pd.read_csv(ruta_tienda2)
df2
```

# Fusión de Datos

```python
df_concat=pd.concat([df, df2], ignore_index=True)
df_concat
```

# Tratamiento de Datos
Realizo tareas de limpieza, filtrado y transformaciones básicas.

Paso 1: Convierto la fecha que es string a datetime

```python
type(df_concat['Fecha'][1])
```

```python
df_concat['Fecha']=pd.to_datetime(df_concat['Fecha'], format='%m/%d/%Y')
```

```python
type(df_concat['Fecha'][1])
```

Paso 2: Ordeno el DataFrame por fecha

```python
df_concat=df_concat.sort_values(by='Fecha')
df_concat
```

Paso 3: Busco mas información de algun dato faltante o duplicados.

```python
df_concat.info()
```

```python
df_concat.describe()
```

```python
df_concat.isna()
```

# Análisis de Ventas

¿Cuál es el producto más vendido?

```python
prod_mas_vendido=df_concat.groupby('Producto')['Cantidad'].sum()
prod_mas_vendido.sort_values(ascending=False).head(1)
```

¿Cuál es el mes con mas ventas?

```python
df_concat['mes']=df_concat['Fecha'].dt.month
```

```python
df_concat
```

```python
mes_mas_ventas=df_concat.groupby('mes')['Cantidad'].sum()
mes_mas_ventas.sort_values(ascending=False).head(1)
```

# Datos Agrupados

```python
agrup_por_categoria=df_concat.groupby('Producto')['Total Venta'].sum()
agrup_por_categoria.sort_values(ascending=False)
```

**Analisis:** El análisis revela que la categoría "Electronics" lidera en el total de ventas, alcanzando una cifra de $244,000. Sin embargo, al considerar la cantidad de unidades vendidas, la categoría de "Alimentos" destaca como la más vendida. Esto sugiere que los productos de la categoría "Electronics" tienen precios significativamente más altos, lo que les permite generar mayores ingresos a pesar de venderse en menor cantidad. Por otro lado, aunque la categoría "Alimentos" registra un mayor volumen de ventas, su precio promedio más bajo resulta en que ocupe el último lugar en el total de ingresos generados.
